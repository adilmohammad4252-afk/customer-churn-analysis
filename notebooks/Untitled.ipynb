📊 Customer Churn Analysis — Insights
🔹 Contract Type

Customers with month-to-month contracts have the highest churn, while customers with long-term contracts (one-year and two-year) show much lower churn.

🔹 Monthly Charges

Customers paying higher monthly charges are more likely to churn compared to those paying lower charges.

🔹 Tenure

Customers with low tenure (new customers) show higher churn, whereas long-term customers are more stable.

🔹 Payment Method

Customers using electronic check have higher churn compared to other payment methods.

🔹 Internet Service

Customers using fiber optic internet tend to churn more than DSL users.
