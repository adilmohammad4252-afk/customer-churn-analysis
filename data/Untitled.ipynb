```python
import pandas as pd

df = pd.read_excel('customer_churn.xlsx')
df.head()
```

```python
df.columns = df.columns.str.strip()
```

```python
df.isnull().sum()
```

```python
df['Total Charges'] = pd.to_numeric(df['Total Charges'], errors='coerce')
```

```python
df = df[df['Total Charges'].notnull()]
```

```python
df['Churn Value']
```

```python
df.drop(['CustomerID', 'Count', 'Lat Long'], axis=1, inplace=True)
```

```python
yes_no_cols = ['Partner', 'Dependents', 'Phone Service', 'Paperless Billing']

for col in yes_no_cols:
    df[col] = df[col].map({'Yes':1, 'No':0})
```

```python
df.info()
df.head()
```

```python
df['Churn Value'].value_counts(normalize=True) * 100
```

```python
pd.crosstab(df['Contract'], df['Churn Value'], normalize='index') * 100
```

```python
pd.crosstab(df['Internet Service'], df['Churn Value'], normalize='index') * 100
```

```python
df.groupby('Churn Value')['Monthly Charges'].mean()
```

```python
df.groupby('Churn Value')['Tenure Months'].mean()
```

```python
pd.crosstab(df['Payment Method'], df['Churn Value'], normalize='index') * 100
```

 Example Insights:
.Customers with month-to-month contracts have the highest churn
.Customers with higher monthly charges are more likely to leave
.Customers with short tenure are at high risk
.Certain payment methods show higher churn
